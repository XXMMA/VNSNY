```python
import pandas as pd
import numpy as np
```

```python
CDdata = pd.read_csv('CDdata.csv')
zoning = pd.read_csv('zoning.csv')
facilities = pd.read_csv('facilities.csv')
```

```python
CDdata.rename(columns = {'borocd':'boroCD'}, inplace = True)
```

```python
from_facilities = ['boroCD','Core Infrastructure and Transportation']
from_zoning = ['boroCD','Residence','Park']
from_CDdata = ['boroCD', 'pop_2010', 'area_sqmi', 'pct_white_nh', 'over65_rate', 'pct_hh_rent_burd', 'pct_served_parks', 'pct_clean_strts', 'crime_per_1000', 'unemployment_cd', 'poverty_rate','count_parks', 'count_hosp_clinic']
data = facilities[from_facilities].merge(zoning[from_zoning], on = 'boroCD').merge(CDdata[from_CDdata], on = 'boroCD')
data.rename(columns = {'Core Infrastructure and Transportation':'transport', 'Residence': 'pct_res', 'area_sqmi' : 'area', 'crime_per_1000':'crime', 'count_parks':'parks','pct_white_nh':'pct_white', 'count_hosp_clinic':'hosp_clinic','pct_hh_rent_burd':'pct_rent_burd', 'Park':'pct_parks', 'over65_rate' : 'pct_over65','rent_burd':'pct_rent_burd','unemployment_cd':'pct_unemployment','poverty_rate':'pct_poverty'}, inplace = True)
convert2float = ['pct_white','pct_over65','pct_rent_burd','pct_served_parks','pct_clean_strts','pct_unemployment','pct_poverty']
data[convert2float] = data[convert2float] / 100
data['pop_dens'] = data['pop_2010'] / data['area']
data['pct_non_res'] = 1 - data['pct_res']
```

```python
data.to_csv('community_district_data.csv', index = False)
```
